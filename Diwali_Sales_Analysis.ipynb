# Exploratory Data Analysis

Gender

```python
df.columns
```

```python
ax=sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

In above graph we clearly show most of the buyers are females and the purchasing power of females are greater than males

# Age

```python
df.columns
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
for i in ax.containers:
    ax.bar_label(i)
```

```python
#Total amount vs age group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age )

 
```

From above graph, we can say most of the buyers are of age between 26-35 yrs females

# State

```python
df.columns
```

```python
#total number of orders from top 10 state
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
#total amount/sales from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

From above graph most of the buyers and total sales/amount are from Uttar Pradesh, Maharashtra and Karnatka respectively

# Merital Status

```python
ax=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(7,5)})
for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(10,7)})
sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')
```

 From above graph most of the buyers are married(female) and they have high purchasing power

# Occupation

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(20,10)})
ax=sns.countplot(x='Occupation',data=df)
for i in ax.containers:
    ax.bar_label(i)
```

```python
#Total amount vs Occupation
sales_age=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=sales_age )
sns.set(rc={'figure.figsize':(20,10)})
 
```

From above graph we can see that most of the buyers are working in IT sector,Hearthcare and Aviation respectively

# Product Category

```python
ax=sns.countplot(x='Product_Category',data=df)
for i in ax.containers:
    ax.bar_label(i)
    
sns.set(rc={'figure.figsize':(20,15)})    
```

```python
sales_product=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount',data=sales_product)
sns.set(rc={'figure.figsize':(26,15)})
```

From above graph most of the sold product are clothes and food and the amount of food is very high

```python
sales_product=df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_ID',y='Amount',data=sales_product)
sns.set(rc={'figure.figsize':(16,10)})
```

```python
sales_product=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(x='Product_ID',y='Orders',data=sales_product)
sns.set(rc={'figure.figsize':(15,10)})
```
